```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 1. Load Input Image
img_path = "car.jpg"          # change to your image filename
cascade_path = "haarcascade_russian_plate_number.xml"  # ensure correct path
```

```python
img = cv2.imread(img_path)

if img is None:
    raise ValueError("Image not found. Check path.")

plt.figure(figsize=(6,6))
plt.title("Input Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# 2. Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5,5), 0)
eq = cv2.equalizeHist(blur)
```

```python
# Show preprocessing steps
plt.figure(figsize=(15,4))
```

```python
plt.subplot(1,3,1)
plt.title("Gray Image")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Blurred")
plt.imshow(blur, cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Histogram Equalized")
plt.imshow(eq, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# -----------------------------
# 4. Load Haar Cascade
# -----------------------------
plate_cascade = cv2.CascadeClassifier(cascade_path)

if plate_cascade.empty():
    raise IOError("Cascade file not loaded. Check path.")


# -----------------------------
# 5. Detect License Plates
# -----------------------------
plates = plate_cascade.detectMultiScale(
    eq,
    scaleFactor=1.1,
    minNeighbors=4,
    minSize=(30, 30)
)

print(f"Detected Plates: {len(plates)}")

# -----------------------------
# 6. Draw Bounding Boxes
# -----------------------------
img_copy = img.copy()

for (x, y, w, h) in plates:
    cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.figure(figsize=(6,6))
plt.title("Detected Plates")
plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# -----------------------------
# 7. Crop & Save Detected Plates
# -----------------------------
count = 1
for (x, y, w, h) in plates:
    plate_roi = img[y:y+h, x:x+w]
    filename = f"plate_{count}.jpg"
    cv2.imwrite(filename, plate_roi)
    print(f"Saved: {filename}")

    # show the cropped plate
    plt.figure(figsize=(4,4))
    plt.title(f"Cropped Plate {count}")
    plt.imshow(cv2.cvtColor(plate_roi, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

    count += 1
```
